```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
%sql
USE CATALOG amazonretail;
USE SCHEMA silver;
```

```python
df_product = spark.table("amazonretail.bronze.bronze_product")
df_customer = spark.table("amazonretail.bronze.bronze_customer")
df_store = spark.table("amazonretail.bronze.bronze_store")
df_transaction = spark.table("amazonretail.bronze.bronze_transaction")
```

```python
df_store = df_store.withColumnRenamed("location", "store_location")
df_customer = df_customer.withColumnRenamed("city", "customer_city")
```

### Joining all the tables

```python
df_final = df_transaction.join(df_product, df_transaction.product_id == df_product.product_id, "left")\
                         .join(df_customer, df_transaction.customer_id == df_customer.customer_id, "left")\
                         .join(df_store, df_transaction.store_id == df_store.store_id, "left")
display(df_final)
```

```python
df_final = df_final.withColumn("sales", col("quantity") * col("price"))
display(df_final)
```

```python
df_final = df_final.drop("product_id", "customer_id", "store_id")
display(df_final)
```

### Writing the joined table to Silver Layer

```python
df_final.write.mode("overwrite")\
              .format("delta")\
              .option("overwriteSchema", "true")\
              .saveAsTable(f"amazonretail.silver.silver_sales")
```

### Writing the table in Delta Format to Silver Container in ADLS

```python
silver_path = 'abfss://silver@amazonprimeadls.dfs.core.windows.net/'

df_final.write.format('delta').mode('overwrite').option('overwriteSchema','true').save(f"{silver_path}silver_sales")
```
